```python
# Aufgabe-25: FIFO-basiertes Tierheimdesign: Im Berliner Tierheim leben nur Katzen und Hunde.
# Und dieses Tierheim arbeitet auf einer „First In First Out – FIFO“-Basis. Der Mensch ist also das „Sensibelste“ der Tiere im Tierheim.
# (nach Ankunfts- oder Einfügungszeit) kann besessen werden. Sie werden Hunde oder Katzen bevorzugen, aber nur diejenigen, die in dieser Art von Tierheim bleiben
# Sie können das älteste/älteste Tier auswählen. Sie können sich das gewünschte Tier nicht aussuchen. Um dieses System zu warten, erstellen Sie die entsprechende Datenstruktur und wenden Sie die folgenden Funktionen an.
# - enqueue()-Funktion; Fügt dem Tierheim (Liste) ein beliebiges Tier hinzu.
# - Funktion dequeue_cat(); Entfernt die älteste Katze aus dem Tierheim (Liste).
# - Funktion dequeue_dog(); Entfernt den ältesten Hund aus dem Tierheim (Liste).
# {Notiz. Das 'älteste/älteste' Tier ist 'das erste Tier, das das Tierheim betritt'}

barinak_dict = {}
kayit_no = 1
def enqueue(tur, hayvan_adi):
    '''''"Katze" und "Hund" zwei Arten von Tieren und ihre Namen als Parameter und mit dem Namen barinak_dict.
     Funktion, die ic ice dictionary mit registration_id hinzufügt'''
    global kayit_no
    barinak_dict.setdefault(tur, {})
    barinak_dict[tur].update({kayit_no:hayvan_adi})   
    kayit_no += 1


enqueue("cat", "Boncuk")
enqueue("cat", "Badem")
enqueue("cat", "zeytin")
enqueue("dog", "Comar")
enqueue("dog", "karabas")
enqueue("cat", "tekir")


# print(barinak_dict) # alle Tiere im Tierheim, bevor Sie Tiere kaufen


def dequeue_cat():
    '''Derjenige, der die älteste Katze im Tierheim aus dem Tierheim geholt hat und
     Funktion, die den letzten Status im Shelter vor und nach dieser Operation druckt'''
    if len(barinak_dict["cat"]) > 0:
        print("Alle Tiere im Tierheim, bevor Sie die Katze bekommen: ", barinak_dict)
        ilk_giren_kedi_id = min(barinak_dict["cat"].keys())
        print("Sie können diese Katze bekommen:" ,barinak_dict["cat"].pop(ilk_giren_kedi_id))
        print("Alle Tiere im Tierheim nach Entnahme der Katze: ", barinak_dict)
        print()
    else:
        print("Leider haben wir keine Katze. Sie können versuchen, einen Hund zu bekommen, wenn Sie möchten.!")
        print()

def dequeue_dog():
    '''der den ältesten Hund im Tierheim aus dem Tierheim geholt hat und
     Funktion, die den letzten Status im Shelter vor und nach dieser Operation druckt'''
    if len(barinak_dict["dog"]) > 0:
        print("Alle Tiere im Tierheim vor Anschaffung eines Hundes: ", barinak_dict)
        ilk_giren_kopek_id = min(barinak_dict["dog"].keys())
        print("Sie können diesen Hund bekommen: " ,barinak_dict["dog"].pop(ilk_giren_kopek_id))
        print("Alle Tiere im Tierheim nach Anschaffung eines Hundes: ", barinak_dict)
        print()
    else:
        print("Leider haben wir keinen Hund. Sie können versuchen, eine Katze zu bekommen, wenn Sie möchten.!")
        print()


dequeue_cat()
dequeue_dog()
dequeue_dog()
dequeue_dog()
```
